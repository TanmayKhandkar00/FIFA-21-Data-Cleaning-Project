**The contract column has three types of values, a year range ('2004 ~ 2021') indicating the the start and end years of the contract, it also has 'Free' as a value indicating that the player is free and not on contract, and a value indicating that the player is on a loan ('Jul 5, 2021 On Loan', here the date is the Loan end date). This column didn't require any transformation. 
From the 'Contract' column, I created three columns derived from it. The 'Contract Start' column containing the year in which the contracted started. The 'Contract End' column containing the end year of the contract. And the 'Contract Length' column containing the length of the contract in years.** 

### Contract column

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
#Separate the contract column 

def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in  contract:
        start_date = np.nan
        end_date = np.nan
        if 'On Loan' in contract:
            contract_length = np.nan
        else:
            contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year 
    return start_date, end_date, contract_length

#Apply the function to the contract column and create new columns

new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc = fifa.columns.get_loc('Contract')+1+i, column = new_cols[i], value = new_data[i])
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)']].sample(5)
```

```python
#Create a new column for the status of the contract of the player

def contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'

#Add contract status column 

fifa.insert(fifa.columns.get_loc('Contract Length(years)')+1, 'Contract status', fifa['Contract'].apply(contract_status))
fifa.sample(5)
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)', 'Contract status']].sample(5)
```

**To standardize the 'Height' column, converted the values to a consistent unit of measurement. The column initially contained a mix of values in centimeters and feet-inches format. Identified the values in feet-inches format and converted them to centimeters. For values already in centimeters, no conversion was needed. This transformation ensured that all height values were expressed uniformly in centimeters, enabling easier analysis and comparison across the dataset. Renamed the column to 'Heigth(cm)' to indicate the unit of measurement.**

### Height

Player's height in centimetres

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
#Function to convert the height in feets and inches to centimetres and removing the 'cm' suffix from the centime

def convert_height(height):
    if 'cm' in height:
        return int(height.strip('cm'))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)

#Apply the function to the height column

fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

```python
# Rename the height column

fifa = fifa.rename(columns = {'Height': 'Heigth(cm)'})
fifa.sample(5)
```

**The 'Weight' column in the dataset contained a mix of values in kilograms (kg) and pounds (lbs). To ensure consistency and facilitate analysis, converted all weight values to kilograms. This allows for accurate and meaningful comparisons across players' weights. Also renamed the column to indicate the unit of measurement to 'Weight(kg)'.**

### Weight

Player's weight in kilograms

```python
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

```python
# Create a function to convert weight in pounds to kilograms and remove the 'kg' suffix

def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        return round(pounds/2.05)

#Apply the function to weight column

fifa['Weight'] = fifa['Weight'].apply(convert_weight)
fifa['Weight'].unique()
        
```

```python
# Rename the weight column

fifa = fifa.rename(columns = {'Weight': 'Weight(kg)'})
fifa.sample(5)
```

**Transformed the 'Joined' and 'Loan Date End' columns which were in string format to datetime format (kept the date format as 'dd-mm-yyyy'). This conversion allows us to leverage datetime functionalities for sorting, filtering, and extracting specific date components. Ultimately, the conversion enhances the usability of the dataset and enables temporal analysis of player careers, transfers, and team dynamics.**

### Joined

Date when the player joined the current club


```python
fifa['Joined'].dtype
```

```python
fifa['Joined'].unique()
```

```python
months_dict = {
    'Jan': 1,
    'Feb': 2,
    'Mar': 3,
    'Apr': 4,
    'May': 5,
    'Jun': 6,
    'Jul': 7,
    'Aug': 8,
    'Sep': 9,
    'Oct': 10,
    'Nov': 11,
    'Dec': 12
}

#Convert the Joined column to date time format
fifa['Joined'] = pd.to_datetime(fifa['Joined'], format='%b %d, %Y')

#applying the dictionary to format the month as 2 digit number

fifa['Joined'] = fifa['Joined'].dt.strftime('%d-%m-%Y').replace(months_dict, regex=True)
fifa['Joined'].unique()
```

**The 'Value', 'Wage' and 'Release Clause' columns indicate amount in Euros about the player's market value, their weekly wage and the release clause in their contract. These values are in srtring format ('€120.5M', '€270K' etc.). Converted these values to numeric format (like 103500000.0 instead of 103.5M), also renamed the column to indicate the currency {'Value(€)', 'Wage(€)', 'Release Clause(€)'}.**

### Value

Estimated market value of the player.

```python
#Convert the 'Value', 'Wage' and 'Release Clasue' columns to numbers
fifa['Value'].dtype
```

```python
fifa['Value'].unique()
```

```python
#Function to convert Value to numeric form

def clean_amount(amount):
    if amount[-1] == 'M':
        return float(amount[1:-1]) * 1000000
    elif amount[-1] == 'K':
        return float(amount[1:-1]) * 1000
    else:
        return  float(amount[1:]) * 1

# #Apply the cleaning function on the value column

fifa['Value'] = fifa['Value'].apply(clean_amount)
fifa['Value'].unique()
```

```python
fifa = fifa.rename(columns = {'Value': 'Value(€)'})
```
